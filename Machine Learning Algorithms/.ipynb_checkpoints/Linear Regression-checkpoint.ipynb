```julia
# Fix column names by replaceing ' ', '-', '/' with '_'
colnames = String[]
for i in String.(names(df))
    push!(colnames,String(replace(replace(replace(strip(i)," " => "_"),"-" => "_"), "/" => "_")))
end

rename!(df, colnames);
```

```julia
#we need to skip missing values or it will throw error
```

```julia
# Box Plot
boxplot(collect(skipmissing(df.Life_expectancy)), title = "Box Plot - Life Expectancy", ylabel = "Life Expectancy (years)", legend = false)
```

```julia
# Outlier removal
first_percentile = percentile(collect(skipmissing(df.Life_expectancy)), 25)
iqr_value = iqr(collect(skipmissing(df.Life_expectancy)))
```

```julia
#since we will be removing it 
#TO BE REVISED
replace!(df.Life_expectancy, Missing=>9999 )
df = df[df.Life_expectancy .<  999,:]
```

```julia
df = df[df.Life_expectancy .>  (first_percentile - 1.5*iqr_value),:];
```

```julia
# Density Plot
density(df.Life_expectancy , title = "Density Plot - Life Expectancy", ylabel = "Frequency", xlabel = "Life Expectancy", legend = false)
```

```julia
# Correlation Analysis
println("Correlation of Life Expectancy with Adult Mortality Rate is ", cor(df.Adult_Mortality,df.Life_expectancy), "\n\n")

# Scatter plot
train_plot = scatter(df.Adult_Mortality,df.Life_expectancy, title = "Scatter Plot Life Expectancy vs Adult Mortality Rate", ylabel = "Life Expectancy", xlabel = "Adult Mortality Rate",legend = false)
```
